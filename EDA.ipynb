# Pie of Avarege years and hands of cars

```python
def extract_col_to_label(col):
    new_dict = dict()
    for val in df_avarege[col]:
        if round(val) not in new_dict:
            new_dict[round(val)] = 1
        else:
            new_dict[round(val)]= new_dict[round(val)]+1
            
    return new_dict
```

```python
year_dict = extract_col_to_label('Avarege year')
pd.options.display.float_format = '{:,.0f}'.format
plt.pie(
    year_dict.values(),
    labels = year_dict.keys(),
    shadow=True,
    autopct='%1.0f%%',
    explode=(0.05, 0.05, 0.05,0.05, 0.05, 0.05, 0.05,0.05),
    wedgeprops = {"edgecolor" : "black",
              'linewidth': 1,
              'antialiased': True})

plt.title('Avarege years of cars')
plt.show()
```

We can see from the pie chart above that most of the second hand cars are from 2014.

```python
hand_dict = extract_col_to_label('Avarege hand')
pd.options.display.float_format = '{:,.0f}'.format
plt.pie(
    hand_dict.values(),
    labels = hand_dict.keys(),
    shadow=True,
    autopct='%1.0f%%',
    explode=(0.05, 0.05,0.05),
    wedgeprops = {"edgecolor" : "black",
              'linewidth': 1,
              'antialiased': True})

plt.title('Avarege hands of cars')
plt.show()
```

We can see from the pie chart above that most of the second hand cars have 2 hands.
