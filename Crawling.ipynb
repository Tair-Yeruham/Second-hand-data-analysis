# Crawling - BeautifulSoup & Selenium

```python
def extract_details_with_soup(url,pages):
    
    time.sleep(10)

    chrome_options = Options()
    chrome_options.add_argument("--kiosk")
    driver = webdriver.Chrome(executable_path = DRIVER_PATH,chrome_options=chrome_options)
    driver.get(url)
    time.sleep(10)
    
    list_year_hand_kilometer = []
    list_price = []
    list_gearbox_color_engineCapacity_area = []

    
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser') # html.parser # lxml
    first_source = soup
    
    size = 1
    while size < pages:
        for car_obj in soup.find_all('div', class_= 'tdKotarotInner'):
            list_year_hand_kilometer.append(car_obj.find('td',class_='tdKoteretRashit').get_text().strip())
            list_price.append(car_obj.find('td',class_='tdPrice').get_text().strip())
            list_gearbox_color_engineCapacity_area.append(car_obj.find('td',class_='tdMoreDetails').get_text().strip())

        size = size +1
        if size == 2:
            driver.find_element_by_css_selector("#paging > a:nth-child("+ str(pages+1)+")").click()
        else:
            driver.find_element_by_css_selector("#paging > a:nth-child("+str(pages+2)+")").click()

        
        time.sleep(10)   
        html = driver.page_source
        soup = BeautifulSoup(html,'html.parser') # source for new page
    
    return pd.DataFrame({"year hand kilimeter":list_year_hand_kilometer,
                         "gearbox color engine capacity area":list_gearbox_color_engineCapacity_area
                         ,"price":list_price})
```
